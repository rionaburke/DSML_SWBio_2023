# Validation of high and low maternal investment lines of the Japanese Quail (_Coturnix japonica_)

**Context:** I will be working with _C. japonica_ selected either from High Maternal Investment (HI) or Low Maternal Investment (LI) breeding lines. This manifests as differing egg sizes in females and asymmetrical testis development in males among other things. 

We will be judging whether the F1 daughters of both HI and LI lines lay eggs of the same type as their mothers (F0) thereby attempting to ascertain the validity of these lines

**Dataset:** Data from this year has not yet been collected as quail are housed in Germany and the SWBio schedule is too busy to allow me to travel. For this, a fabricated dataset based off a previous collection round run by my supervisor reflecting expected values will be used to analyse the contents

# Read in simulated dataset

```python
quail = pd.read_csv("simulated_quail_data_python.csv")
quail
```

# Relationship plot between maternal egg size and the egg size of her daughters

_Plot to see if there is any relationship between maternal egg size and the egg size of her offspring. Blue is the HI maternal line and orange the LI maternal line_

```python
sns.relplot(data = quail, x = quail["AverageEggF1"], y = quail["AverageEggF0"], kind = "line", hue = "MatLine") 
```

# Determining lay rate for F1 offspring

_Lay rate is equal to the number of eggs laid by a bird in a specific observation window (In this case 10 days). Check whether HI lines lay fewer eggs of higher weight as additional validation measure._

```python
sns.catplot(data = quail, x = "MatLine", y = "LayRate", kind = "violin")
```

# Determine mean maternal egg size Vs mean daughter egg size


**1. Average HI maternal egg size (F0)**

```python
high_mat_egg_filter = quail["MatLine"] == "High"
high_mat = quail[high_mat_egg_filter]
high_mat["AverageEggF0"].mean()
```

**2. Average LI maternal egg size (F0)**

```python
low_mat_egg_filter = quail["MatLine"] == "Low"
low_mat = quail[low_mat_egg_filter]
low_mat["AverageEggF0"].mean()
```

**3. Average HI offspring egg size (F1)**

```python
high_mat["AverageEggF1"].mean()
```

**4. Average LI offspring egg size (F1)**

```python
low_mat["AverageEggF1"].mean()
```
